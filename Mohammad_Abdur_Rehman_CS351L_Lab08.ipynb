## **Part 1: Modifying the Neural Network**

### **Step 1: Load and Preprocess the Iris Dataset**
- **Dataset**: The Iris dataset, a classic dataset for classification tasks.
- **Preprocessing**:
  - The features are already scaled, so no further scaling is required.
  - The target labels are one-hot encoded to prepare for multi-class classification.

### **Step 2: Build and Train the Original Network**
- A simple feedforward neural network is constructed with:
  - An input layer matching the number of features.
  - A single hidden layer with 8 neurons and ReLU activation.
  - An output layer with 3 neurons and softmax activation for classification.
- The model is trained using cross-entropy loss and Adam optimizer.

### **Step 3: Modify the Network**
- **Modification**:
  - An additional hidden layer with 16 neurons and ReLU activation is added to increase model capacity.
- The modified network is trained using the same procedure as the original.

### **Step 4: Compare Performance**
- Metrics such as training/validation loss and accuracy are compared between the original and modified models to analyze the impact of the additional layer.

### **Step 1: Load and Preprocess the Wine Dataset**
- **Dataset**: The Wine dataset from Scikit-Learn, used for classification of wine types.
- **Preprocessing**:

### **Step 2: Build and Train the Neural Network**
- **Architecture**:
  - Two hidden layers are used, with 8 and 16 neurons, respectively, and ReLU activation.

```python
# Import necessary libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load and preprocess Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
X_iris = iris.data
y_iris = iris.target.reshape(-1, 1)

# One-hot encode target labels
encoder = OneHotEncoder()
y_iris = encoder.fit_transform(y_iris).toarray()

# Split data into training and testing sets
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42
)

# Build the original network
model_original = Sequential([
    Dense(8, activation='relu', input_shape=(X_train_iris.shape[1],)),
    Dense(3, activation='softmax')
])
model_original.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the original network
history_original = model_original.fit(
    X_train_iris, y_train_iris, epochs=50, validation_split=0.2, verbose=1
)

# Modify the network by adding an extra hidden layer
model_modified = Sequential([
    Dense(8, activation='relu', input_shape=(X_train_iris.shape[1],)),
    Dense(16, activation='relu'),
    Dense(3, activation='softmax')
])
model_modified.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the modified network
history_modified = model_modified.fit(
    X_train_iris, y_train_iris, epochs=50, validation_split=0.2, verbose=1
)

# Load and preprocess Wine dataset
from sklearn.datasets import load_wine
wine = load_wine()
X_wine = wine.data
y_wine = wine.target.reshape(-1, 1)

# Standardize features and one-hot encode target labels
X_wine = StandardScaler().fit_transform(X_wine)
y_wine = encoder.fit_transform(y_wine).toarray()

# Split data into training and testing sets
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(
    X_wine, y_wine, test_size=0.2, random_state=42
)

# Build and train the network for Wine dataset
model_wine = Sequential([
    Dense(8, activation='relu', input_shape=(X_train_wine.shape[1],)),
    Dense(16, activation='relu'),
    Dense(3, activation='softmax')
])
model_wine.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history_wine = model_wine.fit(
    X_train_wine, y_train_wine, epochs=50, validation_split=0.2, verbose=1
)

# Function to visualize training results
def plot_training_results(history, title):
    # Plot training and validation loss
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot training and validation accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Visualize results for the original, modified, and Wine dataset models
plot_training_results(history_original, "Original Model (Iris Dataset)")
plot_training_results(history_modified, "Modified Model (Iris Dataset)")
plot_training_results(history_wine, "Model (Wine Dataset)")

# Function to generate and plot confusion matrix
def plot_confusion_matrix(model, X_test, y_test, encoder, dataset_name):
    # Predict and convert to class labels
    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_true = np.argmax(y_test, axis=1)

    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Display confusion matrix
    plt.figure(figsize=(8, 6))
    plt.title(f'Confusion Matrix - {dataset_name}')
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=encoder.categories_[0]).plot(
        cmap=plt.cm.Blues, ax=plt.gca()
    )
    plt.show()

# Plot confusion matrix for Iris dataset (original and modified models)
plot_confusion_matrix(model_original, X_test_iris, y_test_iris, encoder, "Iris Dataset (Original Model)")
plot_confusion_matrix(model_modified, X_test_iris, y_test_iris, encoder, "Iris Dataset (Modified Model)")

# Plot confusion matrix for Wine dataset model
plot_confusion_matrix(model_wine, X_test_wine, y_test_wine, encoder, "Wine Dataset")

```
