#### **Scenario**
The scenario involves placing knights on a 3x3 chessboard such that no two knights can attack each other. This is achieved by representing the chessboard as a graph where each vertex corresponds to a square on the board, and edges represent valid knight moves. The goal is to use graph coloring to ensure that no two connected vertices (squares) share the same color, which corresponds to placing knights in a way that they do not threaten each other.

#### **Steps and Functionality**

1. **Define the Graph with Knight Moves**
   - The graph `G` is created with 9 vertices (0 to 8), representing the squares on a 3x3 chessboard.
   - Edges are added to the graph to represent valid knight moves. Each edge connects two vertices that a knight can move between in one move.

2. **Visualize the Uncolored Graph**
   - The function `visualize_uncolored_graph()` visualizes the initial state of the chessboard with no knights placed. Nodes are displayed in white, and edges are black.

3. **Apply a Greedy Coloring Algorithm**
   - The function `greedy_coloring(G)` applies a greedy coloring algorithm to the graph. This algorithm assigns colors to the vertices such that no two adjacent vertices share the same color. The colors represent different knights.

4. **Visualize the Colored Graph**
   - The function `visualize_colored_graph(colors)` visualizes the chessboard after applying the greedy coloring algorithm. Each node is colored based on the assigned color, representing the placement of knights.

5. **Display the Uncolored Graph**
   - The script calls `visualize_uncolored_graph()` to display the initial state of the chessboard.

6. **Color the Graph and Visualize**
   - The script calls `greedy_coloring(G)` to color the graph and then `visualize_colored_graph(colors)` to display the colored chessboard with knights placed.



```python
# Step 1: Define the graph with knight moves on a 3x3 chessboard (9 vertices)
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Define the graph with 9 vertices (0 to 8) and edges representing knight moves
G = nx.Graph()

# These edges represent the valid moves a knight can make on a 3x3 board
# Each edge is a move from one square to another
knight_moves = [(0, 7), (0, 5), (1, 6), (1, 8), (2, 7), (2, 5), (3, 8), (3, 6), (4, 5), (4, 7), (4, 8)]
G.add_edges_from(knight_moves)

# Step 2: Visualize the uncolored graph (initial chessboard with no knights)
def visualize_uncolored_graph():
    plt.figure(figsize=(6, 6))
    pos = nx.spring_layout(G, seed=42)  # Use spring layout for spacing nodes
    node_colors = ['#ffffff'] * len(G.nodes)  # White color for uncolored nodes
    edge_colors = 'black'
    node_border_color = 'black'

    nx.draw(G, pos, with_labels=True, node_color=node_colors, font_weight='bold',
            node_size=2000, font_size=16, font_color='black', edge_color=edge_colors,
            linewidths=2, node_shape='o', edgecolors=node_border_color)

    plt.title("Chessboard (3x3) with Knight Moves (Uncolored)")
    plt.show()

# Step 3: Apply a greedy coloring algorithm to place knights
def greedy_coloring(G):
    # Use a greedy algorithm to color the graph
    colors = nx.coloring.greedy_color(G, strategy="largest_first")
    return colors

# Step 4: Visualize the colored chessboard with knights
def visualize_colored_graph(colors):
    plt.figure(figsize=(6, 6))
    pos = nx.spring_layout(G, seed=42)
    # Define a color map for the knight placement (colors correspond to knights)
    knight_colors = ['#ff9999', '#99ccff', '#ffff99', '#66ff66', '#ffcc66', '#ff6666', '#66cccc', '#cc99ff', '#ff9966']
    node_colors = [knight_colors[colors[node]] for node in G.nodes]  # Assign colors based on the result of greedy coloring
    edge_colors = 'black'
    node_border_color = 'black'

    nx.draw(G, pos, with_labels=True, node_color=node_colors, font_weight='bold',
            node_size=2000, font_size=16, font_color='black', edge_color=edge_colors,
            linewidths=2, node_shape='o', edgecolors=node_border_color)

    plt.title("Knight Placement using Graph Coloring")
    plt.show()

# Step 5: Display the uncolored graph (initial state)
visualize_uncolored_graph()

# Step 6: Color the graph using the greedy algorithm and visualize
colors = greedy_coloring(G)
visualize_colored_graph(colors)
```

## **layer vs AI Chess Game (Graph Representation)**

This Python script simulates a game where a player and an AI take turns placing knights on a 3x3 chessboard. The game uses the NetworkX library to model the chessboard and knight moves as a graph, and Matplotlib to visualize the board state after each move.

### **Steps and Functions:**

1. **Define the Chessboard and Knight Moves**:
   - The chessboard is represented as a graph where each node is a position on the board.
   - Edges represent valid knight moves on a 3x3 board.

2. **Visualize the Board**:
   - The `visualize_board` function uses Matplotlib to display the current state of the board.
   - Knights placed by the player and AI are shown in different colors.

3. **Initialize the Board State**:
   - The `initialize_board` function sets up an empty board with no knights placed.

4. **Player Move**:
   - The `player_move` function prompts the player to input a position to place their knight.
   - Ensures the move is valid and updates the board state.

5. **AI Move**:
   - The `ai_move` function uses a greedy graph coloring algorithm to place AI knights.
   - Ensures AI knights are placed in valid positions.

6. **Check for Game Over**:
   - The `game_over` function checks if all positions on the board are occupied, indicating the end of the game.

7. **Game Loop**:
   - The `play_game` function runs the main game loop, alternating between player and AI moves until the game is over.




```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Step 1: Define the chessboard with knight moves (3x3 board)
G = nx.Graph()

# Knight moves on a 3x3 board
knight_moves = [(0, 7), (0, 5), (1, 6), (1, 8), (2, 7), (2, 5), (3, 8), (3, 6), (4, 5), (4, 7), (4, 8)]
G.add_edges_from(knight_moves)

# Visualize the board at any point
def visualize_board(current_positions, player_moves):
    plt.figure(figsize=(6, 6))
    pos = nx.spring_layout(G, seed=42)

    # Define colors for AI knights and player knights
    knight_colors = ['#99ccff' if current_positions[node] == 'AI' else '#ff9999' if current_positions[node] == 'Player' else '#ffffff' for node in G.nodes]

    edge_colors = 'black'
    node_border_color = 'black'

    nx.draw(G, pos, with_labels=True, node_color=knight_colors, font_weight='bold',
            node_size=2000, font_size=16, font_color='black', edge_color=edge_colors,
            linewidths=2, node_shape='o', edgecolors=node_border_color)

    plt.title(f"Chessboard after Player Move: {player_moves}")
    plt.show()

# Step 2: Initialize the board state
def initialize_board():
    # Empty positions at the start (None means no knight at that position)
    return {i: None for i in range(9)}

# Step 3: Player move function (input from player)
def player_move(board_state):
    while True:
        try:
            move = int(input("Enter a position (0-8) to place your knight: "))
            if 0 <= move < 9 and board_state[move] is None:
                board_state[move] = 'Player'
                return move
            else:
                print("Invalid move! Try again.")
        except ValueError:
            print("Please enter a valid number.")

# Step 4: AI move function using graph coloring
def ai_move(board_state):
    # Use greedy coloring to assign AI knights safely
    available_positions = [node for node, val in board_state.items() if val is None]
    if not available_positions:
        return None

    # Use graph coloring to determine valid placement
    colors = nx.coloring.greedy_color(G, strategy="largest_first")
    for position in available_positions:
        if colors[position] == 0:  # AI uses 'color 0' for its moves
            board_state[position] = 'AI'
            return position

    # If no valid moves are found (rare case), pick randomly
    move = random.choice(available_positions)
    board_state[move] = 'AI'
    return move

# Step 5: Check if the game is over (no valid moves)
def game_over(board_state):
    return all(val is not None for val in board_state.values())

# Step 6: Run the game loop
def play_game():
    board_state = initialize_board()
    player_moves = []

    while not game_over(board_state):
        # Player makes a move
        print("Your turn:")
        player_move_pos = player_move(board_state)
        player_moves.append(player_move_pos)
        visualize_board(board_state, player_moves)

        if game_over(board_state):
            print("Game over! No more valid moves.")
            break

        # AI makes its move
        print("AI's turn:")
        ai_move_pos = ai_move(board_state)
        if ai_move_pos is None:
            print("AI cannot make a move!")
            break
        print(f"AI placed a knight at position {ai_move_pos}.")
        visualize_board(board_state, player_moves)

    print("Game finished!")

# Step 7: Start the game
play_game()
```

## **Player vs AI Chess Game (Grid Representation)**

This Python script simulates a simplified chess game where a player and an AI take turns placing knights on a 3x3 grid. The game uses the NetworkX library to model the knight moves as a graph and employs a greedy graph coloring algorithm for the AI's moves.

### **Steps and Functions:**

1. **Create a Graph for Knight Moves**:
   - The `generate_knight_graph` function creates a graph where nodes represent positions on the 3x3 grid and edges represent valid knight moves.

2. **Initialize the Chessboard**:
   - The `initialize_board` function sets up an empty 3x3 grid represented by a list of lists.

3. **Display the Chessboard**:
   - The `display_board` function prints the current state of the chessboard to the console.

4. **Get Available Moves**:
   - The `get_available_moves` function returns a list of empty positions on the board where a knight can be placed.

5. **Player Move**:
   - The `player_move` function prompts the player to input a row and column to place their knight. It ensures the move is valid and updates the board.

6. **AI Move Using Graph Coloring**:
   - The `ai_move` function uses a greedy graph coloring algorithm to determine the AI's move. It places the AI's knight in a valid position and updates the board.

7. **Main Game Loop**:
   - The `play_game` function runs the main game loop, alternating between player and AI moves until no more valid moves are available.



```python
import networkx as nx
import random

# Step 1: Create a graph for knight moves
def generate_knight_graph():
    G = nx.Graph()
    knight_moves = [(0, 7), (0, 5), (1, 6), (1, 8), (2, 7), (2, 5), (3, 8), (3, 6), (4, 5), (4, 7), (4, 8)]
    G.add_edges_from(knight_moves)
    return G

# Step 2: Initialize a chessboard (3x3 grid for simplicity)
def initialize_board(size=3):
    return [['.' for _ in range(size)] for _ in range(size)]

# Step 3: Display the chessboard
def display_board(board):
    for row in board:
        print(" ".join(row))
    print()

# Step 4: Get available moves on the board
def get_available_moves(board):
    available_moves = []
    for i in range(len(board)):
        for j in range(len(board[i])):
            if board[i][j] == '.':  # Empty spot
                available_moves.append((i, j))
    return available_moves

# Step 5: Player move
def player_move(board):
    while True:
        try:
            row, col = map(int, input("Enter your move as row and column (e.g., 0 1): ").split())
            if board[row][col] == '.':
                board[row][col] = 'P'
                break
            else:
                print("Invalid move! The position is already taken.")
        except (ValueError, IndexError):
            print("Invalid input! Please enter valid row and column numbers.")

# Step 6: AI move using graph coloring
def ai_move(board, G):
    available_moves = get_available_moves(board)
    if not available_moves:
        return None

    # Use graph coloring to select AI move
    colors = nx.coloring.greedy_color(G, strategy="largest_first")
    for (row, col) in available_moves:
        node = row * 3 + col
        if colors.get(node, 1) == 0:  # AI uses 'color 0' for its moves
            board[row][col] = 'A'
            return row, col

    # If no valid moves, pick randomly
    row, col = random.choice(available_moves)
    board[row][col] = 'A'
    return row, col

# Step 7: Main game loop
def play_game():
    # Initialize graph and board
    G = generate_knight_graph()
    board = initialize_board()

    print("Welcome to the Player vs AI Chess Game (Grid Representation)")
    display_board(board)

    while True:
        # Player's turn
        print("Your turn:")
        player_move(board)
        display_board(board)

        # Check for available moves for the AI
        if not get_available_moves(board):
            print("No more moves available! Game over.")
            break

        # AI's turn
        print("AI's turn:")
        ai_move(board, G)
        display_board(board)

        # Check for available moves for the player
        if not get_available_moves(board):
            print("No more moves available! Game over.")
            break

# Run the game
play_game()
```
